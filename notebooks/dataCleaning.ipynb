## Missing Values

```python
df.isna().sum()
```

```python
# nan as percentage of total
missingValues = df.isna().sum()[df.isnull().sum()>0] / len(df)
missingValues.sort_values(ascending=False)
```

```python
# before missing value column removal
df.shape
```

```python
# removing columns with more than 50% missing values
df = df.dropna(thresh=len(df)*0.5, axis=1) # 41 columns removed
df.shape
```

```python
# still a few columns with a lot of missing values
missingValues = df.isna().sum()[df.isnull().sum()>0] / len(df)
missingValues.sort_values(ascending=False)
```

**For simplicity now we will also drop the columns with more than 30% missing values**

However, a better solution would be to impute the missing values. They could be imputed with the mean, median or mode of the column. Or a more complex method could be used, such as k-nearest neighbors or a regression model.

```python
# removing columns with more than 30% missing values
df = df.dropna(thresh=len(df)*0.7, axis=1) # 9 columns removed
df.shape
```

**dropping missing rows**

```python
df = df.dropna()
df.info(verbose=True, show_counts=True)
```

## Loan Types

```python
df['NAME_CONTRACT_TYPE'].value_counts()
```

**Filtering Out Revolving Loans**

```python
df = df[df['NAME_CONTRACT_TYPE'] == 'Cash loans']
df.info(verbose=True, show_counts=True)
```

```python
df = df.drop(['NAME_CONTRACT_TYPE'], axis = 1)
```

## Encoding

```python
columns = df.columns
stringColumns = []

for column in columns:
    if df[column].dtype == 'object':
        stringColumns.append(column)
    else:
        continue

print(stringColumns)
```

```python
# value counts for each string column

for column in stringColumns:
    valueCounts = df[column].value_counts()
    print(valueCounts)
    print('-----------------------------------')
```

```python
twoValues = []
moreThanTwoValues = []

for column in stringColumns:
    if len(df[column].value_counts()) <= 2:
        twoValues.append(column)
    else:
        moreThanTwoValues.append(column)
print(len(twoValues), ' string columns with 2 unique values')
print(len(moreThanTwoValues), ' string columns with more than 2 unique values')
```

### One-Hot Encoding

```python
# One-Hot Encoding Columns with 2 unique values
twoValues
```

```python
# encode 'CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY' columns
df = pd.get_dummies(df, columns=twoValues, drop_first=True)
df.head(3)
```

```python
moreThanTwoValues
```

```python
# drop ORGANIZATION_TYPE as it has too many unique values
df = df.drop(['ORGANIZATION_TYPE'], axis = 1)
```

```python
# One-Hot Encoding Columns with more than 2 unique values
# TODO: label encode education type and family status
moreThanTwoValues.remove('ORGANIZATION_TYPE')
df = pd.get_dummies(df, columns=moreThanTwoValues, drop_first=True)
df.info(verbose=True, show_counts=True)
```
